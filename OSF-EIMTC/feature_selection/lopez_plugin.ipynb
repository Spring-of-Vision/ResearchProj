```python
import pandas as pd
import random
import numpy as np
```

```python
def process_packet_data(df):
    # Function to process each row's 'Data' column
    def process_packets(packets):
        # Keep only the first 32 packets
        truncated_packets = packets[:32]
        
        # Remove the timestamp from each packet
        processed_packets = [[packet[1], packet[2], packet[3]] for packet in truncated_packets]
        
        return processed_packets

    # Apply the function to each row's 'Data' column
    df['Data'] = df['Data'].apply(process_packets)
    
    return df
```

```python
def process_packet_data_to_numpy(df):
    # Function to process each row's 'Data' column
    def process_packets_to_numpy(packets):
        truncated_packets = packets[:32]
        #print(truncated_packets)
        processed_packets = [[float(packet[1]), int(packet[2]), int(packet[3])] for packet in truncated_packets]
        return np.array(processed_packets)
    
    # Apply the function to each row's 'Data' column
    df['Data'] = df['Data'].apply(process_packets_to_numpy)
    
    return df
```

```python
# Create DataFrame
#df = pd.DataFrame(data)

# Load DataFrame
df = pd.read_parquet('quic_text.parquet')
```

```python
df
```

```python
# Process the DataFrame to convert 'Data' to NumPy arrays
processed_df = process_packet_data(df)
processed_df
```

```python
processed_df.drop(columns=['num_of_packets'], inplace= True)
processed_df
```

```python
# Save Features to CSV file
processed_df.to_csv("lopez_features.csv", index=False)

# Save DataFrame to a Parquet file using pyarrow
processed_df.to_parquet('lopez_features.parquet', engine='pyarrow')

# (parquet doesn't work with 2d array)
```
